<h3 style="font-family: SimSun, 宋体, serif; font-size: 12pt; line-height: 1.5; margin: 0; text-align: justify;">1. 相较于传统手工操作交易的方法，量化交易有哪些优势？</h3>
<p style="font-family: SimSun, 宋体, serif; font-size: 10.5pt; line-height: 1.5; margin: 0; text-align: justify;">
量化交易把交易思想转化为可计算、可回测、可重复执行的规则。相比手工交易，它的优势主要包括：
</p>
<ul style="font-family: SimSun, 宋体, serif; font-size: 10.5pt; line-height: 1.5; margin: 0; text-align: justify;">
<li>执行纪律更强，减少情绪化追涨杀跌</li>
<li>可以快速处理大量股票、行情和财务数据</li>
<li>策略可以通过历史数据回测，提前评估收益、回撤和风险</li>
<li>交易信号和下单过程可以自动化，提高速度和一致性</li>
<li>便于持续优化和风险控制，例如设置止损、仓位限制和组合分散</li>
</ul>
</div>

<h3 style="font-family: SimSun, 宋体, serif; font-size: 12pt; line-height: 1.5; margin: 0; text-align: justify;">2. 基本概念解释</h3>
<p style="font-family: SimSun, 宋体, serif; font-size: 10.5pt; line-height: 1.5; margin: 0; text-align: justify;">
<strong>K线：</strong>K线用开盘价、最高价、最低价、收盘价描述一个交易周期的价格变化，常用于观察趋势、波动和多空力量。
</p>
<p style="font-family: SimSun, 宋体, serif; font-size: 10.5pt; line-height: 1.5; margin: 0; text-align: justify;">
<strong>基本面：</strong>基本面关注公司盈利、资产负债、行业景气、宏观经济、政策等影响股票长期价值的因素。
</p>
<p style="font-family: SimSun, 宋体, serif; font-size: 10.5pt; line-height: 1.5; margin: 0; text-align: justify;">
<strong>技术面：</strong>技术面主要研究价格、成交量和技术指标，试图从市场行为中判断趋势、支撑阻力和买卖时机。
</p>
</div>

<div style="font-family: SimSun, 宋体, serif; font-size: 10.5pt; line-height: 1.5; margin: 0; text-align: justify;">
<h3 style="font-family: SimSun, 宋体, serif; font-size: 12pt; line-height: 1.5; margin: 0; text-align: justify;">3. Python 获取过去一年交易数据、画收盘价曲线并保存 CSV</h3>
</div>

```python
import os
from datetime import datetime, timedelta
import tushare as ts
import pandas as pd
import matplotlib.pyplot as plt

# 设置中文字体
plt.rcParams['font.sans-serif'] = ['SimSun']
plt.rcParams['axes.unicode_minus'] = False

# 获取 Tushare Token
token = os.getenv('TUSHARE_TOKEN')
assert token, '请先设置环境变量 TUSHARE_TOKEN'
ts.set_token(token)
pro = ts.pro_api()

# 设置股票代码和时间范围
ts_code = '000001.SZ'
end_date = datetime.today().strftime('%Y%m%d')
start_date = (datetime.today() - timedelta(days=365)).strftime('%Y%m%d')

# 获取数据
print(f'正在获取 {ts_code} 从 {start_date} 到 {end_date} 的数据...')
df = pro.daily(ts_code=ts_code, start_date=start_date, end_date=end_date)
df = df.sort_values('trade_date').reset_index(drop=True)

# 保存数据
df.to_csv('../data/task1_daily_data.csv', index=False, encoding='utf-8-sig')

# 显示数据基本信息
print(f'数据获取成功，共 {len(df)} 条记录')
print(f'时间范围：{df["trade_date"].min()} 至 {df["trade_date"].max()}')
print('\n数据前5行：')
display(df.head())

# 绘制收盘价走势图
plt.figure(figsize=(12, 6))
plt.plot(pd.to_datetime(df['trade_date']), df['close'], label='收盘价', color='#1266d6', linewidth=1.5)
plt.title(f'{ts_code} 每日收盘价走势', fontsize=14, fontweight='bold')
plt.xlabel('日期', fontsize=12)
plt.ylabel('收盘价', fontsize=12)
plt.legend(fontsize=10)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
